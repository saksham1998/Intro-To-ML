# Applying Machine Learning Algo.

```python
data.head()
```

```python
X = data.iloc[:,:-1].values
```

```python
y = data.iloc[:,-1].values
```

```python
X[:3]
```

```python
y[:3]
```

```python
from sklearn.preprocessing import MinMaxScaler
X_scaler = MinMaxScaler()
X = X_scaler.fit_transform(X)
```

```python
X[:5]
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=101,stratify=y)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
# Applying Logistic Regression

from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
```

```python
from sklearn.model_selection import cross_val_score
```

```python
accuracy_comp = {}
recall_comp = {}
```

```python
def accuracy(model,name):
    acc_mean = cross_val_score(model,X_train,y_train,cv=10,scoring='accuracy').mean()
    accuracy_comp[name] = acc_mean
```

```python
def recall(model,name):
    recall_mean = cross_val_score(model,X_train,y_train,cv=10,scoring='recall_weighted').mean()
    recall_comp[name] = recall_mean
```

```python
accuracy(log_model,'LogisticRegression')    
```

```python
recall(log_model,'LogisticRegression')
```

```python
accuracy_comp
```

```python
recall_comp
```

```python
log_model.fit(X_train,y_train)
```

```python
y_pred = log_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,f1_score,recall_score
print(classification_report(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
# SVM
```

```python
from sklearn.svm import SVC
```

```python
svc_model = SVC()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
Cs = [0.001, 0.01, 0.1, 1, 10]
gammas = [0.001, 0.01, 0.1, 1]
kernels = ['linear','poly','rbf']
param_grid = {'C': Cs, 'gamma' : gammas,'kernel':kernels}
grid = GridSearchCV(svc_model,param_grid,cv=10,scoring='recall_weighted')
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```

```python
grid.best_score_
```

```python
accuracy_comp['SVC'] = grid.best_score_
```

```python
svc_model2 = SVC(C=1,gamma=0.001,kernel='linear')
```

```python
svc_model2.fit(X_train,y_train)
```

```python
y_pred = svc_model2.predict(X_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```

```python
recall_score(y_test,y_pred)
```

```python
recall_comp['SVC'] = recall_score(y_test,y_pred)
```
