```python
plt.figure(figsize=(8,4))
sns.heatmap(data.isnull(),cmap='cividis',cbar=False,yticklabels=False)
plt.title('Checking any null values')
plt.xlabel('Features')
```

i.e no null values present in data

```python
# Dropping the invaluable features , ('RowNumber','CustomerId','Surname')

data.drop(['RowNumber','CustomerId','Surname'],axis=1,inplace=True)
```

# Region Demographics

```python

plt.figure(figsize=(9,5))
sns.countplot(data['Geography'])
plt.title(' Region Demographics ')
plt.xlabel(' Countries ')
plt.ylabel('Count')
```

i.e major population of bank customers live in france about 5000 and others are about equally divided in Sapin and Germany

```python
print('Percentage of Customer in France {:10}%'.format(len(data[data['Geography']=='France'])/100))
print('Percentage of Customer in Spain  {:10}%'.format(len(data[data['Geography']=='Spain'])/100))
print('Percentage of Customer in Germany{:10}%'.format(len(data[data['Geography']=='Germany'])/100))
```

```python
plt.style.use('fivethirtyeight')
plt.pie([50.14,24.77,25.01],labels=['France','Spain','Germany'],explode=[0.01,0.02,0.03],startangle=90,shadow=True)
plt.title('Region Demographics')
```

# Credit Score Evaluation

```python
# Credit Score Evaluation
import warnings
warnings.filterwarnings('ignore')
plt.style.use('fivethirtyeight')
sns.distplot(data['CreditScore'],kde=False)
plt.title('Credit Score Evaluation',color='black')
plt.xlabel('Score')
plt.ylabel('Count')
```

Credit Score Chart
<img src="files/credit_score.png">

```python
print('% of Customers with Excellent Credit Score {:10}%'.format(len(data[data['CreditScore']>=800])/100))
print('% of Customers with Very Good Credit Score {:10}%'.format(len(data[(data['CreditScore']>=740)&(data['CreditScore']<800)])/100))
print('% of Customers with Average Credit Score   {:10}%'.format(len(data[(data['CreditScore']>=670)&(data['CreditScore']<740)])/100))
print('% of Customers with Low Credit Score       {:10}%'.format(len(data[(data['CreditScore']>=580)&(data['CreditScore']<670)])/100))
print('% of Customers with Very Low Credit Score  {:10}%'.format(len(data[(data['CreditScore']>=300)&(data['CreditScore']<580)])/100))
```

From Above we can clearly see that about 57% of cutomers have credit score ranging from 300-670 

```python

plt.style.use('ggplot')
plt.pie([6.65,12.24,24.28,33.31,23.62],labels=['Excellet','Very Good','Average','Bad','Very Low'],startangle=90,shadow=True,explode=[0.01,0.04,0.03,0.03,0.02],colors=['red','orange','blue','purple','cyan'])
plt.title('Credit Score Evaluation')
```

```python
data['CreditScore'].mean()
```

So,in general average creadit score of bank customer falls in 'Low - Average' Section 

```python
#Regional Check of Credit Score

print(data[data['Geography']=='France']['CreditScore'].mean())
print(data[data['Geography']=='Spain']['CreditScore'].mean())
print(data[data['Geography']=='Germany']['CreditScore'].mean())
```

As we know that only 18.79%(6.55+12.24) customers falls under Very Good to Excellent category

Checking These Customers Region Wise

```python
print('France {}'.format(len(data[(data['Geography']=='France')&(data['CreditScore']>740)])))
print('Germany {}'.format(len(data[(data['Geography']=='Germany')&(data['CreditScore']>740)])))
print('Spain {}'.format(len(data[(data['Geography']=='Spain')&(data['CreditScore']>740)])))
```

```python
plt.style.use('ggplot')
sns.boxplot(data['Geography'],data['CreditScore'])
plt.title('Region Wise Credit-Score Evaluation')
```

# Gender Classification
