```python
data['diagnosis'] = data['diagnosis'].map({'M':1,'B':0})
```

```python
data.head()
```

```python
X = data.iloc[:,1:].values
```

```python
y = data.iloc[:,0].values
```

```python
X[0:5]
```

```python
y[0:5]
```

```python
from sklearn.cross_validation import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=101,stratify=y)
```

```python
X_train[:5]
```

```python
y_train[:5]
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score

confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train,y_train)
```

```python
k_y_pred = knn.predict(X_test)
```

```python
confusion_matrix(y_test,k_y_pred)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
```

```python
dtc.fit(X_train,y_train)
```

```python
d_y_pred = dtc.predict(X_test)
```

```python
confusion_matrix(y_test,d_y_pred)
```

```python
print('Logistic Regression Accuracy : {}'.format(accuracy_score(y_test,y_pred)))
print('K Nearest Neighbors Accuracy : {}'.format(accuracy_score(y_test,k_y_pred)))
print('Decision Tree Accuracy : {}'.format(accuracy_score(y_test,d_y_pred)))
```
