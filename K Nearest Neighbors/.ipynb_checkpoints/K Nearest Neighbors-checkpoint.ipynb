```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn import datasets
iris = datasets.load_iris()
```

```python
type(iris)
```

```python
iris.keys()
```

```python
X = iris['data']
```

```python
y = iris['target']
```

```python
iris['target_names'] # 0 :setosa, 1 :versicolor , 2 :virginica
```

```python
df = pd.DataFrame(X,columns = iris['feature_names'])
```

```python
df.head()
```

```python
pd.plotting.scatter_matrix(df,c=y,figsize=(10,8))
# sns.pairplot(df)
```

```python
sns.pairplot(df)
```

```python
from sklearn.cross_validation import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=101,stratify=y)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
confusion_matrix(y_pred,y_test)
```

```python
accuracy_score(y_pred,y_test)
```

```python
#It shows that are model has accuracy of 92%.Great!!!!!
```

```python
val = model.predict([[10.2,0.5,0.8,10.9],[1.2,0.5,0.8,1.9],[5.6, 2.7, 4.2, 1.3]])
for i in val:
    if i==0:
        print('Setosa')
    elif i==1:
        print('versicolor')
    else:
        print('virginca')
```
